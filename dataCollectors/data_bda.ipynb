```python
import os
import pandas as pd
```

```python
directory_path = '/Users/asal/Desktop/bda_viatges/airbnb_data'
```

```python
file_names = [f for f in os.listdir(directory_path) if f.endswith('.csv')]
```

```python
file_names
```

```python
combined_df = pd.DataFrame()

for file_name in file_names:
    
    city, day_type = file_name.split('_')
    day_type = day_type.replace('.csv', '').capitalize() 
    
    
    file_path = os.path.join(directory_path, file_name)
    df = pd.read_csv(file_path, index_col=0)

    df['city'] = city.capitalize() 
    df['day_type'] = day_type

    combined_df = pd.concat([combined_df, df])
combined_df.shape

```

```python
combined_df
```

```python
combined_df.to_csv('/Users/asal/Desktop/data_airbnb.csv', index=False)
```
