```python
from FlightRadar24 import FlightRadar24API
import json
fr_api = FlightRadar24API()
```

## Obtenir aeroports de les ciutats que ens interessen

```python
airports = fr_api.get_airports()
```

```python
airports
```

```python
# Llista de ciutats
ciutats = ['Amsterdam','Athens','Barcelona','Berlin','Budapest','Lisbon','London','Paris','Rome','Vienna']
```

```python
import re
regex = r'\((.*?)\)'
aeroports_inclosos = []

for ap in airports:
    for ciutat in ciutats:
        if ciutat in str(ap):
            sigles = re.findall(regex,str(ap))[0]
            aeroports_inclosos.append(ap)
            break
```

```python
aeroports_inclosos
```

```python
def parseig_aeroports(lista_aeropuertos):
    aeropuertos_info = {}
    regex = r'\((.*?)\) .*? Latitude: ([-+]?\d+\.\d+) - Longitude: ([-+]?\d+\.\d+)'
    
    for aeropuerto in lista_aeropuertos:
        matches = re.findall(regex, str(aeropuerto))
        if matches:
            siglas = matches[0][0]
            latitud = float(matches[0][1])
            longitud = float(matches[0][2])
            aeropuertos_info[siglas] = {'Latitud': latitud, 'Longitud': longitud}
    
    return aeropuertos_info
```

```python
airports_info = parseig_aeroports(aeroports_inclosos)
```

```python
def save_json(dicc,file):
    with open(file,'w') as f:
        f.write(json.dumps(dicc))
```

```python
save_json(airports_info,'airports_info.json')
```
